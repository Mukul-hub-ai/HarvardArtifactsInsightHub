```python
import requests

API_KEY = "9ca1111c-6a05-45dc-817e-7bab5939072a"
url = "https://api.harvardartmuseums.org/object"

all_records = []

for page in range(1, 31):
    params = {
        "apikey": API_KEY,
        "size": 100,
        "page": page,
        "classification": "Coins"
    }

    response = requests.get(url, params=params)
    clean_data = response.json()

    records = clean_data.get("records", [])
    if not records:
        print(f"No records found on page {page}")
        break

    all_records.extend(records)
    print(f"Page {page} fetched. Total records so far: {len(all_records)}")

print(f"\nTotal {len(all_records)} records fetched.")

```

```python
all_records[0]
```

```python
metadata = []
media = []
colors = []

for i in all_records:
    metadata.append(dict(
        id = i.get('id'),
        title = i.get('title', 'unknown'),
        culture = i.get('culture', 'unknown'),
        period = i.get('period', 'unknown'),
        century = i.get('century', 'unknown'),
        medium = i.get('medium', 'unknown'),
        dimensions = i.get('dimensions', 'unknown'),
        description = i.get('description', 'No description'),
        department = i.get('department', 'unknown'),
        classification = i.get('classification', 'unknown'),
        accessionyear = i.get('accessionyear', 'unknown'),
        accessionmethod = i.get('accessionmethod', 'unknown')
    ))

    media.append(dict(
        objid = i.get('objectid'),
        imagecount = i.get('imagecount'),
        mediacount = i.get('mediacount'),
        colorcount = i.get('colorcount'),
        rank = i.get('rank'),
        datebegin = i.get('datebegin'),
        dateend = i.get('dateend')           
    ))

    color_details = i.get('colors')
    if color_details:
        for j in color_details:
            colors.append(dict(
                objid = i.get('objectid'),
                color = j.get('color'),
                spectrum = j.get('spectrum'),
                hue = j.get('hue'),
                percent = j.get('percent'),
                css3 = j.get('css3')
            ))
```

```python
# Establish the connection
connection = mysql.connector.connect(
        host="localhost",
        port=3306,
        user='root',
        password="root",
        database="project_1" )

cursor = connection.cursor()

```

```python
cursor.execute("ALTER TABLE metadata modify column century varchar(100)")
connection.commit()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS metadata(   
    id INT,
    title VARCHAR(255),
    culture VARCHAR(100),
    period VARCHAR(100),
    century VARCHAR(20),
    medium TEXT,
    dimensions TEXT,
    description TEXT,
    department VARCHAR(100),
    classification VARCHAR(100),
    accessionyear VARCHAR(10),
    accessionmethod VARCHAR(100)
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS media (
    objid INT,
    imagecount INT,
    mediacount INT,
    colorcount INT,
    `rank` INT,
    datebegin INT,
    dateend INT
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS colors (
    objid INT,
    color VARCHAR(50),
    spectrum VARCHAR(100),
    hue VARCHAR(50),
    ercent FLOAT,
    css3 VARCHAR(50)
)
""")

connection.commit()    # this is used for permanently saving the data 
```

```python
insert_query = """
INSERT INTO metadata (id, title, culture, period, century, medium, dimensions,
description, department, classification, accessionyear, accessionmethod)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"""

values = []
for row in metadata:
    values.append((
        row['id'], row['title'], row['culture'], row['period'], row['century'],
        row['medium'], row['dimensions'], row['description'], row['department'],
        row['classification'], row['accessionyear'], row['accessionmethod']
    ))

cursor.executemany(insert_query, values)
connection.commit()
```

```python
cursor.execute("DESCRIBE metadata;")
for col in cursor.fetchall():
    print(col)
```

```python
media_insert_query = """
INSERT INTO media (objid, imagecount, mediacount, colorcount, `rank`, datebegin, dateend)
VALUES (%s, %s, %s, %s, %s, %s, %s)
"""

media_values = []
for row in media:
    media_values.append((
        row['objid'], row['imagecount'], row['mediacount'], row['colorcount'],
        row['rank'], row['datebegin'], row['dateend']
    ))

cursor.executemany(media_insert_query, media_values)
connection.commit()
print("media table data inserted successfully.")
```

```python
colors_insert_query = """
INSERT INTO colors (objid, color, spectrum, hue, ercent, css3)
VALUES (%s, %s, %s, %s, %s, %s)
"""

colors_values = []
for row in colors:
    colors_values.append((
        row['objid'], row['color'], row['spectrum'], row['hue'],
        row['percent'], row['css3']
    ))

cursor.executemany(colors_insert_query, colors_values)
connection.commit()
print("colors table data inserted successfully.")
```
